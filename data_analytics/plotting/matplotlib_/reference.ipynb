```python
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [14, 9]
import numpy as np
```

```python
fig, ax = plt.subplots()

x = np.linspace(0, 2, 100)

ax.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
ax.plot(x, x**2, label='quadratic')  # etc.
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Simple Plot")
ax.legend()

plt.show() # to suppress last object string
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [1, 4, 9], 'b--')
ax.plot([1, 2, 3], [1, 4, 9], 'r*')
# colors: r, g, b, y, w, k, 
# marker styles: , . o x _ + *
# line styles: - -- :
fig.set_size_inches(10.0, 6.0)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2,1)

ax1.plot([1, 2, 3], [1, 4, 9], 'b--')
ax1.plot([1, 2, 3], [1, 4, 9], 'r*')

ax2.plot([1, 2, 3], [1, 4, 9], 'b--')
ax2.plot([1, 2, 3], [1, 4, 9], 'r*')

plt.show()
```
