```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from scipy import stats
```

```python
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [14, 9]
```

```python
number_of_obs = 1000
# my_seed = 42
# np.random.seed(my_seed)
mvnorm = stats.multivariate_normal(mean=[0, 0], cov=[[1., 0.9],  
                                                     [0.9, 1.]])
X_norm = mvnorm.rvs(number_of_obs)
X = stats.norm().cdf(X_norm)
X1 = X[:,0]
X2 = X[:,1]
```

```python
np.corrcoef(X1,X2)
```

```python
plt.plot(list(range(len(X1))), X1, 'o')
```

```python
plt.hist(X1, bins=100)
plt.show()
```

```python
plt.plot(X1, X2, 'o')
```

```python
residuals = stats.norm(loc = 0, scale = 1).rvs(number_of_obs)
y = 30 * X1 - 10 * X2 + residuals
```

```python
lr = LinearRegression()
lr.fit(X,y)
lr.coef_
```

```python
plt.plot(X2, y, 'o')
```
