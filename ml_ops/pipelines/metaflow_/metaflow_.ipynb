```python
from metaflow import FlowSpec, step, NBRunner

class SimpleFlow(FlowSpec):
    
    @step
    def start(self):
        print("Starting pipeline...")
        self.data = [1, 2, 3, 4, 5]
        self.next(self.process)
    
    @step
    def process(self):
        print(f"Processing data: {self.data}")
        self.result = sum(self.data)
        self.next(self.end)
    
    @step
    def end(self):
        print(f"Pipeline complete! Result: {self.result}")

# Run the flow directly in the notebook
NBRunner(SimpleFlow).nbrun()
```
